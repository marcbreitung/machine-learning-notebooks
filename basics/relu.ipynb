# The Rectified Linear Unit (ReLU)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-5, 5, num=100)
y = X
```

```python
plt.plot(X,y)
plt.vlines([0], -5, 5, colors='r')
plt.grid(True)
```

```python
r_y = np.maximum(0, y)
```

```python
plt.plot(X, r_y)
plt.vlines([0], -1, 5, colors='r')
plt.grid(True)
```
